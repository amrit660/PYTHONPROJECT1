```python
print("=== Shopping List Manager ===")

# Using individual variables (no lists allowed)
item1 = ""
item2 = ""
item3 = ""
item4 = ""
item5 = ""
count = 0

while True:
    print("\n1. Add item")
    print("2. View items")
    print("3. Remove item")
    print("4. Count items")
    print("5. Exit")

    choice = input("Choose option: ")

    if choice == "1":
        new_item = input("Enter item: ")
        if item1 == "":
            item1 = new_item
        elif item2 == "":
            item2 = new_item
        elif item3 == "":
            item3 = new_item
        elif item4 == "":
            item4 = new_item
        elif item5 == "":
            item5 = new_item
        else:
            print("List full (max 5 items).")
        print("Item added!")

    elif choice == "2":
        print("Current items:")
        if item1 != "":
            print(item1)
        if item2 != "":
            print(item2)
        if item3 != "":
            print(item3)
        if item4 != "":
            print(item4)
        if item5 != "":
            print(item5)

    elif choice == "3":
        remove_item = input("Enter item to remove: ")
        if item1 == remove_item:
            item1 = ""
            print("Item removed.")
        elif item2 == remove_item:
            item2 = ""
            print("Item removed.")
        elif item3 == remove_item:
            item3 = ""
            print("Item removed.")
        elif item4 == remove_item:
            item4 = ""
            print("Item removed.")
        elif item5 == remove_item:
            item5 = ""
            print("Item removed.")
        else:
            print("Item not found.")

    elif choice == "4":
        count = 0
        if item1 != "":
            count += 1
        if item2 != "":
            count += 1
        if item3 != "":
            count += 1
        if item4 != "":
            count += 1
        if item5 != "":
            count += 1
        print("Total items:", count)

    elif choice == "5":
        print("Goodbye!")
        break

    else:
        print("Invalid option. Try again.")
```
